#### We need to filter out data, in which have city have palyed at least 5 innings.
#### In this dataset, if we play a match, match have two innings in every format, bating match_id are different.
#### We have T-20 WC data.


```python
#6 balls in over, , 20 over in  per inning , 5 inning played 

(6*20)*5
```

```python
cities = df['city'].value_counts()
cities
```

```python
#need 5 inning in a city data, so balls all are deliver in each city more tahn or equal 600 deliveries

five_inning_cities = cities[cities >= 600]
five_inning_cities
```

```python
five_inning_cities = cities[cities >= 600].index.tolist()
five_inning_cities
```

```python
# now only those data comes where 600 or above balls are delivery in each city

df = df[df['city'].isin(five_inning_cities)]
df
```

```python
df.groupby('match_id')['runs'].cumsum().iloc[115:175]
```

```python
# find ut current score and make new column

df['current_score'] = df.groupby('match_id')['runs'].cumsum()
df['current_score'] 
```

```python
df
```

```python
#make a new column 'over' from Column 'ball', for check which over is going

df['over'] = df['ball'].astype(int)
df
```

```python
#make a new column 'ball_number' on RE on the base of which ball in over

df['ball_number'] = df['ball'].astype(str).str.extract(r'\d\.(\d)').astype(int)
df
```

```python
df['total_deliveries'] = (df['over'] *6) + df['ball_number']
df
```

```python
df.info()
```

```python
#Number of matches which have extra balls deliver
extra_delivery = df[df['total_deliveries'] >120]
extra_delivery
```

```python
extra_delivery['match_id'].unique()
```

```python
# Assuming extra_delivery is your DataFrame
unique_match_ids = extra_delivery['match_id'].nunique()
print("Number of matches which have extra balls deliver:", unique_match_ids)
```

```python
#find out how many balls are left

df['ball_left'] = 120 - df['total_deliveries']
df
```

```python
# In column "ball_left", we will se extraa ball are in -ve shown in numbers, so avoid -ve number change to 0

df['ball_left'].mask(df['ball_left'] <0 , 0 , inplace = True)
df
```

```python
#list of dismiss players names

df['player_dismissed'].unique()
```

```python
#now remove palyer name when he will out replace into 1

df['player_dismissed'] = df['player_dismissed'].mask(df['player_dismissed'] != '0' , 1).astype(int)

df
```

```python
df['match_id'].unique()
```

```python
# Number of unique match IDs / How many innings played

unique_match_ids = len(df['match_id'].unique())
print("Number of unique match IDs / How many innings played:", unique_match_ids)
```

```python
# find how many players are out in an innings
df["player_dismissed"] = df.groupby("match_id")["player_dismissed"].cumsum()
df
```

```python
#make new column 'Wickets_left'

df['wickets_left'] = 10 - df['player_dismissed']
df
```

```python
#make a new column 'current run rate'

df['crr'] = (df['current_score'] * 6) / df['total_deliveries'] 
                                         
df
```

```python
#find final score in individual innings
total_runs_an_inning = df.groupby(['match_id'])['runs'].sum().reset_index()
total_runs_an_inning
```

```python
#merge the column in dataset

df = df.merge(total_runs_an_inning , on = 'match_id')
df
```

```python
#make a column of 'last five over' runs 

df['last_five_overs'] = df.groupby('match_id')['runs_x'].rolling(window = 30).sum().values.tolist()
df
```

```python
final_df = df[['batting_team', 'bowling_team','city' ,'ball' , 'ball_left', 'wickets_left','last_five_overs',
                'runs_y']]
final_df
```

```python
#now we wiLl feature engineering (F.E.) in column 'run_y', it show how many rung in that inning

x = final_df.drop(columns = ['runs_y']) #in x-axis its drop but in y-axis individual values comes from dataset
y = final_df['runs_y']

```

```python
#train and test the dataset using sklearn model

from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error

```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse=False),
    ['batting_team', 'bowling_team' , 'city']
)], remainder='passthrough')
```

```python
pipe = Pipeline(steps = [
    ('step1' ,trf),
    ('step2' ,StandardScaler()),
    ('step3' , XGBRegressor (n_estimators = 1000 , learning_rate =0.2 ,max_depth = 12 , random_state = 1))
    
])
```

```python
pipe
```

```python
#find how much % model is correct and 2nd values show that runs error in mean

pipe.fit(x_train , y_train)

y_pred = pipe.predict(x_test)

print(r2_score(y_test , y_pred))

print(mean_absolute_error(y_test , y_pred))
```

```python
pickle.dump(pipe , open('pipe.pkl' , 'wb'))
```
